## Transform decision

```python
df = pd.read_csv('asylum-decisions.csv')
df.head()
```

```python
df = df[['Year','Country of asylum','Recognized','Complementary protection','Rejected','Otherwise closed','Total decisions']]
```

```python
# df = df.set_index('Year')
df = df.groupby(by=['Country of asylum']).sum()
df = df.reset_index()
df = df.drop('Year',axis=1)
df.head()
```

```python
df['Percentage of recognized'] = df['Recognized']/df['Total decisions']
df['Percentage of Complementary protection'] = df['Complementary protection']/df['Total decisions']
df['Percentage of Rejected'] = df['Rejected']/df['Total decisions']
df['Percentage of closed'] = df['Otherwise closed']/df['Total decisions']
df
```

```python
# df.to_csv('asylum-decisions_transformed1.csv')
df.head()
```

```python
df = pd.melt(df,id_vars='Country of asylum',value_vars = df.columns[1:])
df.head()
# 
```

```python
df.to_csv('asylum-decisions_transformed3.csv')
```
